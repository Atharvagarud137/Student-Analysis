```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("student_scores.csv")
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
df.isnull().sum()
```

# Drop unnamed column

```python
df = df.drop("Unnamed: 0", axis=1)
df.head()
```

# Change weekly study hours column

```python
df["WklyStudyHours"] = df["WklyStudyHours"].str.replace("05-Oct", "5-10")
df.head()
```

# Gender Distribution

```python
plt.figure(figsize=(5,5))
ax = sns.countplot(data = df, x = "Gender")
ax.bar_label(ax.containers[0])
plt.title("Gender Distribution")
plt.show()
```

From the above chart, we analysed that the number of females is higher than the number of males

```python
gb = df.groupby("ParentEduc").agg({"MathScore":'mean', "ReadingScore":'mean', "WritingScore":'mean'})
print(gb)
```

```python
plt.figure(figsize=(5,5))
sns.heatmap(gb, annot=True)
plt.title("Effect of Parent Education on Student's Scores")
plt.show()
```

From the above chart, we have concluded that the education of parents has a good impact on the maths, reading and writing score of the students.

```python
gb1 = df.groupby("ParentMaritalStatus").agg({"MathScore":'mean', "ReadingScore":'mean', "WritingScore":'mean'})
print(gb1)
```

```python
plt.figure(figsize=(5,5))
sns.heatmap(gb1, annot=True)
plt.title("Effect of Parent Marital Status on Student's Scores")
plt.show()
```

From the above chart, we have concluded that the marital status of the parents does not have any major effect on the maths, reading and writing score of the students.

# Distribution of Ethnic Groups

```python
groupA = df.loc[(df["EthnicGroup"] == "group A")].count()
groupB = df.loc[(df["EthnicGroup"] == "group B")].count()
groupC = df.loc[(df["EthnicGroup"] == "group C")].count()
groupD = df.loc[(df["EthnicGroup"] == "group D")].count()
groupE = df.loc[(df["EthnicGroup"] == "group E")].count()

l = ["group A", "group B", "group C", "group D", "group E"]
list = [groupA["EthnicGroup"], groupB["EthnicGroup"], groupC["EthnicGroup"], groupD["EthnicGroup"], groupE["EthnicGroup"]]
plt.pie(list, labels=l, autopct="%1.2f%%")
plt.title("Distibution of Ethnic Groups")
plt.show()
```
